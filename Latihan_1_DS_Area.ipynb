```python
df = pd.read_csv('HR_comma_sep.csv')
```

```python
df.duplicated().sum()
```

```python
for col in df.select_dtypes(include=np.number):
  plt.hist(df[col])
  plt.title(col)
  plt.show()
```

#Mean, modus dan median

```python


# Assuming df is your DataFrame
for col in df.select_dtypes(include=np.number).columns:
    rerata = df[col].mean()
    modus = df[col].mode().iloc[0]  # mode() returns a Series, so take the first value
    median = df[col].median()

    # Values to plot
    x_labels = ['Rata-rata', 'Modus', 'Median']
    x_values = [rerata, modus, median]

    # Plotting
    plt.bar(x_labels, x_values, color='red')
    plt.title(f"{col} Statistik")  # Add column name to title
    plt.ylabel('Nilai')

    # Adding value labels
    for i, v in enumerate(x_values):
        plt.text(i, v, f"{v:.2f}", ha='center', va='bottom')  # Format with 2 decimal places

    plt.show()
```

#Vertical Boxplot

```python
for col in df.select_dtypes(include=np.number):
  plt.figure(figsize=(5, 7))  # Adjust the figure size as needed
  sns.boxplot(y=df[col])
  plt.show()
```

#Horizontal Boxplot

```python
for col in df.select_dtypes(include=np.number):
  plt.figure(figsize=(7, 5))  # Adjust the figure size as needed
  sns.boxplot(x=df[col])
  plt.show()
```
